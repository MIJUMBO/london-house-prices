**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Here's where you should write your function to calculate our ratio by dividing the Average_Price of 1998 by Average_Price of 1998 after filtering through our dataset :
def create_price_ratio(d):
    price_1998 = d.loc[d['Year'] == 1998, 'Average_price'].iloc[0]
    price_2018 = d.loc[d['Year'] == 2018, 'Average_price'].iloc[0]
    ratio = [price_1998/price_2018]
    return ratio
```

```python
#  Test out the function by calling one of the London Boroughs with the following argument:

df_grp[df_grp['London_Borough']=='Barking & Dagenham']
create_price_ratio(df_grp[df_grp['London_Borough']=='Barking & Dagenham'])
```

```python
# Create an empty dictionary to store results
final = {}

# Loop through each borough
for borough_name in df_grp['London_Borough'].unique():
    # Filter rows for that borough
    borough_data = df_grp[df_grp['London_Borough'] == borough_name]
    
    # Calculate the price ratio using our function
    ratio = create_price_ratio(borough_data)
    
    # Store the result in the dictionary
    final[borough_name] = ratio

# Print the results
print(final)
```

```python
#Coverting 

df_ratio = pd.DataFrame(final)
```

```python
# Call the head() method on this variable to check it out. 
df_ratio.head()
```

```python
# All we need to do now is transpose it, and reset the index! 
df_ratio_T = df_ratio.transpose()
df_ratios = df_ratio_T.reset_index()
df_ratios.head()
```

```python
# Renaming the 'index' column as 'London_Borough', and the '0' column to 'Ratio_2018'.
df_ratios = df_ratios.rename(columns={'index':'Borough', 0:'Ratio_2018'})
df_ratios.head()
```

```python
# Lets create another column to show the actual growth rate of each of London_borough between 1998 to 2018
df_ratios['Growth_%'] = (1 / df_ratios['Ratio_2018'] - 1) * 100

df_ratios.head()
```

```python
# To answer our question we are going to sort the Growth rate (Growth_%) in descending order for the top 15 boroughs.
# Next we create a variable called top15 and sort our df_ratios by the top 15 locations with the highest growth rate 

top15 = df_ratios.sort_values(by='Growth_%',ascending= False).head(15)
top15
```

```python
# Let's plot the boroughs that have seen the greatest changes in price.
# Make a variable called ax and assign it the result of filtering top15 on 'Borough' and 'Growth_%', then calling plot() using a horizontal bar chart using pandas
ax = top15[['Borough', 'Growth_%']].plot(
    kind='barh',
    x='Borough',
    y='Growth_%',
    color='teal',
    figsize=(10,6),
    legend=True,
)

# Customize titles and labels
ax.set_title('House Price Growth (1998–2018) for London Borough', fontsize=12)
ax.set_xlabel('Growth Percentage (%)')
ax.set_ylabel('London Borough')
ax.invert_yaxis()

# Add labels automatically (no loop needed)
ax.bar_label(ax.containers[0], fmt='%.0f%%')

# Display the plot
plt.show()
```
